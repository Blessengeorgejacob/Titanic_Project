```python
dummy=pd.get_dummies(data[['Name','Sex','Embarked','Ticket','Cabin']],drop_first=True)
dummy
```

```python
new_data=pd.concat([data,dummy],axis=1)
new_data
```

```python
new_data=new_data.drop(['PassengerId','Name','Sex','Embarked','Ticket','Cabin'],axis=1)
new_data
```

```python
new_data.isna().sum()
```

```python
new_data.dtypes
```

```python
new_data['Age']=new_data['Age'].fillna(new_data['Age'].mode()[0])
```

```python
new_data.isna().sum()
```

```python
new_data
```

```python
x=new_data.drop(['Survived'],axis=1)
x
```

```python
y=new_data['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model3=LinearRegression()
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
y_pred
```

```python
x1=pd.DataFrame({"Actual_value":y_test,"predicted_value":y_pred})
x1
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
error
```

```python
from sklearn.metrics import mean_absolute_percentage_error
per=mean_absolute_percentage_error(y_test,y_pred)
per
```

```python
from sklearn.metrics import mean_squared_error
squ=mean_squared_error(y_test,y_pred)
squ
```

```python
from sklearn.metrics import mean_squared_error
sq=np.sqrt(squ)
sq
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
